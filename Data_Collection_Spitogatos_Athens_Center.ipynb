```python
#print(soup_object.prettify())
```

```python
def sales_property(website, col):
    record = []
    for i in website:
        driver.get(i)
        soup_object = soup(driver.page_source, 'html.parser')
        details= soup_object.find_all("div", {"class": "margin-top-small"} ,  {"id": "searchDetailsListings"})
        list_1 = details[2].find_all("div", {"class":"searchListing_details__inner clearfix"})

        
        for item in list_1:
            try:
                #Price of house
                price = float(item.find('span',"text xbig bold padding-right").text.replace('€ ', '').replace(',',''))
                
                #Square meters house
                square_meter_house = float(item.find('span',"padding-right text medium").text.replace('\n','').replace('m2', ''))
                
                #Number of Bedrooms
                bedrooms = float(item.find('span',"text medium").text.replace('\n',' ').replace('bedrooms', '').replace('bedroom', ''))
                
                #Description
                description = item.find("div", {"class":"margin-bottom-small text color dark_grey show-screen"}).text.replace('\n',' ')
                
                #Location
                location = item.find("div", {"class":"text color dark_grey margin-bottom-small listingLocation"}).text
                
                #House ID
                id_home = item.find_all("a")
                id_house = id_home[1].get('href').split('/',4)[-1]
                
                
            
            except AttributeError:
                n = math.nan
                price = n
                square_meter_house = n
                bedrooms = n
                description = n
                location = n
                house_ID = n
    
            results = [price,id_house, square_meter_house, bedrooms, location, description]
            record.append(results)
       
    
    #time.sleep(10)
    
    
    df = pd.DataFrame(record, columns = col)
    df["price_per_sqrmeter"] = df["Price"] / df["Square meters house"]
    
    return df
```
